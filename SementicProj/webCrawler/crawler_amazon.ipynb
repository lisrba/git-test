```python
base_url = "http://www.amazon.com/"
with open('BLACK-DECKER-CHV1410L-Lithium-Cordless.csv', 'wb') as csvfile:
    writer = csv.writer(csvfile, delimiter=',',
                        quotechar='"', quoting=csv.QUOTE_MINIMAL)
    total_num_pages=188
    pageNum=1
    while(pageNum<=total_num_pages):

        url="BLACK-DECKER-CHV1410L-Lithium-Cordless/product-reviews/B006LXOJC0/ref=cm_cr_getr_d_paging_btm_"+str(pageNum)+"?ie=UTF8&pageNumber="+str(pageNum)+"&reviewerType=all_reviews&pageSize=1000"
        uri=base_url+url
        print uri

        try:
            with requests.Session() as session:
                r=session.get(uri)
                print "status=",r.status_code
                if (r.status_code==200):

                    soup = BeautifulSoup(r.content,"lxml")
                    sleep(1)

                    count=0
                    for elem in soup.select(".review-text"):
                        count+=1
                        print pageNum,count,'\n'
                        #print elem.text,'\n\n'
                        writer.writerow([pageNum,count,elem.text.encode("utf-8")])
                    pageNum=pageNum+1
                else:
                    print "Connection failed. Reconnecting..."
        except:
            print "Connection error or something wrong. URL=",uri
```
